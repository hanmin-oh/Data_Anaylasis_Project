```python
import warnings
warnings.filterwarnings(action='ignore')
%config Computer.use_jedi = False
import numpy as np
import pandas as pd
```

```python
width = pd.read_csv('./data/동북부면적.csv')
width
```

```python
width['면적'] = width['면적'] * 0.000001
width
```

```python
width['면적'] = width['면적'].round(2)
width
```

```python
# 면적을 내림차순으로 정렬
width_sorted = width.sort_values(by='면적', ascending=False)
# 상위 20개 데이터를 선택
top_20_width = width_sorted.head(20)
top_20_width
```
