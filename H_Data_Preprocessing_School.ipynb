```python
high_info = pd.read_csv('./data/고등학교 정보.csv')
high_info
```

```python
high_info.drop(columns=['시도교육청', '지역교육청','학교급코드', '지역', '정보공시 학교코드','설립구분', '학교특성',
       '분교여부', '설립유형', '주야구분', '개교기념일', '설립일', '법정동코드', '주소내역', '상세주소내역', '우편번호', '학교도로명 우편번호',
        '전화번호', '팩스번호', '홈페이지 주소', '남녀공학 구분'], inplace=True)
high_info
```

```python
# 카카오맵 API Key
kakao_api_key = '146c824df784569973cb17777ddd3301'

high_info['행정동'] = None
high_info['시군구'] = None
# 카카오맵 API를 호출하여 주소별 위도, 경도 및 행정동명을 얻습니다.
for index, row in high_info.iterrows():
    latitude = row['위도']
    longitude = row['경도']
    
    url = f'https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?x={longitude}&y={latitude}'
    headers = {
        'Authorization': f'KakaoAK {kakao_api_key}'
    }
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        data = response.json()
        if data.get('documents'):
            administrative_district1 = data['documents'][1]['region_3depth_name']
            administrative_district2 = data['documents'][1]['region_2depth_name']
            
            high_info.at[index, '행정동'] = administrative_district1
            high_info.at[index, '시군구'] = administrative_district2
            

# 결과를 출력합니다.
high_info
```

```python
high_info.drop(columns=['학교도로명 주소', '학교도로명 상세주소', '위도', '경도'], inplace=True)
high_info
```

```python
high_info.to_csv('./data/행정동시군구학교명.csv')
high_info
```

```python
# '강북구, '도봉구', '노원구' , '중랑구', '동대문구' 선택
selected_districts = ['강북구', '도봉구', '노원구', '성북구', '중랑구', '동대문구']
filtered_df = high_info[high_info['시군구'].isin(selected_districts)]
filtered_df
```

```python
grouped = filtered_df.groupby(['시군구', '행정동']).size().reset_index(name='학교수')
grouped
```

```python
width = pd.read_csv('./data/h동북부면적.csv')
width
```

```python
merged_df = pd.merge(grouped, width, on=['시군구', '행정동'], how='inner')
merged_df
```

```python
merged_df['면적(㎢)당 학교 수'] = ( merged_df['학교수'] / merged_df['면적']).round(2)
merged_df
```

```python
merged_df.drop(columns=['학교수', '면적'], inplace=True)
merged_df
```
