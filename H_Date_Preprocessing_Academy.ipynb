```python
import warnings
warnings.filterwarnings(action='ignore')
%config Computer.use_jedi = False
import numpy as np
import pandas as pd
```

```python
shop = pd.read_csv('./data/shop_2022_seoul.csv')
shop.head()
```

```python
value_columns = ['상권업종대분류명', '상권업종중분류명',  '상권업종소분류명',  '시도명', '시군구명', 
                 '행정동명','법정동명', '지번주소', '도로명주소',  '경도', '위도']
shop_col = shop[value_columns]
shop_col
```

```python
del shop
```

```python
shop_col = shop_col[shop_col.상권업종대분류명 == '교육']
shop_col
```

# 입시 학원

```python
shop_col_edu = shop_col[shop_col.상권업종소분류명 == '입시·교과학원']
shop_col_edu
```

```python
# 모델링에 필요한 서울특별시, 시군구, 법정동명을 제외하고 칼럼을 삭제한다. 
shop_col_edu.drop(['상권업종대분류명', '상권업종중분류명', '법정동명', '지번주소', '도로명주소', '경도', '위도'], axis=1, inplace=True)
shop_col_edu
```

```python
shop_col_edu = shop_col_edu.rename(columns={'시군구명': '시군구', '행정동명': '행정동'})
shop_col_edu
```

```python
# '강북구, '도봉구', '노원구' , '중랑구', '동대문구' 선택
selected_districts = ['강북구', '도봉구', '노원구', '성북구', '중랑구', '동대문구']
filtered_df = shop_col_edu[shop_col_edu['시군구'].isin(selected_districts)]
filtered_df
```

```python
# '시군구'과 '법정동'을 기준으로 그룹화하고 '행정동'의 데이터 수를 카운트
grouped = filtered_df.groupby(['시군구', '행정동']).size().reset_index(name='학원 수')
grouped
```

```python
grouped.to_csv('./data/h학원수.csv')
grouped
```

```python
grouped[grouped['행정동'] == '중계본동']
```

```python
# 조인
merged_df = pd.merge(grouped, width, on=['시군구', '행정동'], how='inner')
merged_df
```

```python
merged_df['면적(㎢)당 학원 수'] =  ( merged_df['학원 수'] / merged_df['면적'] ).round(2)
merged_df
```
