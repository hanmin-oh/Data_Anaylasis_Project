```python
import warnings
warnings.filterwarnings(action='ignore')
%config Computer.use_jedi = False
import numpy as np
import pandas as pd
```

```python
house = pd.read_csv('./data/동북부집값.csv')
house = house.drop(columns=['Unnamed: 0'], axis=1)
# house['면적(㎢)당금액'] = house['면적(㎡)당금액'] * 0.001
# house.drop(columns=['면적(㎡)당금액'], inplace=True)
house
```

```python
school = pd.read_csv('./data/동북부학교수.csv')
school = school.drop(columns=['Unnamed: 0'], axis=1)
school
```

```python
merge = pd.merge(house, school, on=['시군구', '법정동'], how='left')
merge
```

```python
academy = pd.read_csv('./data/동북부학원수.csv')
academy = academy.drop(columns=['Unnamed: 0'], axis=1)
academy
```

```python
merge = pd.merge(merge, academy, on=['시군구', '법정동'], how='left')
merge
```

```python
subway = pd.read_csv('./data/동북부지하철수.csv')
subway = subway.drop(columns=['Unnamed: 0'], axis=1)
subway
```

```python
merge = pd.merge(merge, subway, on=['시군구', '법정동'], how='left')
merge
```

```python
# '면적(㎢)당 금액' 칼럼을 '면적(㎡)당금액'을 맨 뒤로 이동
merge = merge[[col for col in merge if col != '면적(㎡)당금액'] + ['면적(㎡)당금액']]

# 변경된 데이터프레임을 출력하거나 저장할 수 있습니다.
merge
```

```python
# NaN 값을 0으로 채웁니다.
merge['면적(㎢)당 역 수'].fillna(0, inplace=True)
merge['면적(㎢)당 학교 수'].fillna(0, inplace=True)
merge['면적(㎢)당 학원 수'].fillna(0, inplace=True)
merge
```

### 상관관계 계산

```python
df = merge
```

```python
df.columns = ['시군구', '법정동', 'school', 'academy', 'subway', 'target']
```

```python
X1 = merge.subway.values
X2 = merge.school.values
X3 = merge.academy.values
Y = df.target.values
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'D2Coding'
plt.rcParams['font.size'] = 15
```

```python
plt.scatter(X1, Y, alpha=0.5)
plt.title('집값과 지하철역의 상관관계')
plt.xlabel('지하철 역')
plt.ylabel('집값')
plt.show()
```

```python
plt.scatter(X2, Y, alpha=0.5)
plt.title('집값과 학교의 상관관계')
plt.xlabel('학교')
plt.ylabel('집값')
plt.show()
```

```python
plt.scatter(X3, Y, alpha=0.5)
plt.title('집값과 학원의 상관관계')
plt.xlabel('학원')
plt.ylabel('집값')
plt.show()
```
