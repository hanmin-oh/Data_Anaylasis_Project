```python
import warnings
warnings.filterwarnings(action='ignore')
%config Computer.use_jedi = False
import numpy as np
import pandas as pd
import requests
```

```python
high_info = pd.read_csv('./data/고등학교 정보.csv')
high_info
```

```python
high_info.columns
```

```python
high_info.drop(columns=['시도교육청', '지역교육청','학교급코드', '지역', '정보공시 학교코드','설립구분', '학교특성',
       '분교여부', '설립유형', '주야구분', '개교기념일', '설립일', '법정동코드', '주소내역', '상세주소내역', '우편번호', '학교도로명 우편번호',
        '전화번호', '팩스번호', '홈페이지 주소', '남녀공학 구분'], inplace=True)
high_info
```

```python
# '학교도로명 주소' 열에서 '송파구' 추출하여 '시군구' 열에 할당
high_info['시군구'] = high_info['학교도로명 주소'].apply(lambda x: x.split()[1] if len(x.split()) > 2 else '')

# 결과 확인
high_info
```

```python
# '강북구, '도봉구', '노원구' , '중랑구', '동대문구' 선택
selected_districts = ['강북구', '도봉구', '노원구', '성북구', '중랑구', '동대문구']
filtered_df = high_info[high_info['시군구'].isin(selected_districts)]
filtered_df
```

```python
filtered_df.drop(columns=['학교도로명 주소', '학교도로명 상세주소'], inplace=True)
filtered_df
```

```python
filtered_df.to_csv('./data/w동북부학교.csv')
filtered_df
```
